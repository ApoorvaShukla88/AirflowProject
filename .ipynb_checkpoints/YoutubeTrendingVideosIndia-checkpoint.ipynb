```python
import pandas as pd
from datetime import date
import calendar
import json
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/amishra/DEV/DataEngineering.Labs.AirflowProject/DataEngg-Airflow/INvideos.csv')
df['month'] = df.apply(lambda row: calendar.month_name[int(row.trending_date.split(".")[2])], axis=1)
df
```

```python
df1 = pd.read_json('/Users/amishra/DEV/DataEngineering.Labs.AirflowProject/DataEngg-Airflow/IN_category_id.json')
df1
```

```python
data = None
with open('/Users/amishra/DEV/DataEngineering.Labs.AirflowProject/DataEngg-Airflow/IN_category_id.json') as f:
    data = json.load(f)

category_list = list()
for c in data['items']:
    category_list.append([c['id'],c['snippet']['title']])

df2 = pd.DataFrame(category_list, columns = ['category_id', 'category_title'])
df2
```

```python
df.groupby('month').count()
```

```python
df.groupby('category_id').count()
```

```python
df12 = df2.astype({'category_id': int})
df12.info()

```

```python
merged_inner = pd.merge(df, df12, on='category_id')
merged_inner
```

```python
df13 = merged_inner.groupby(['month','category_title']).count()
df13
```

```python
#df13 = pd.read_csv('/Users/amishra/DEV/DataEngineering.Labs.AirflowProject/DataEngg-Airflow/dataclean/INvideos_aggregated.csv')
#df13
df14 = df13.iloc[: , 0:1]
df14.columns.str.strip()
df14
```

```python
df14.plot.line()
```

```python
fig, ax = plt.subplots(figsize=(15,7))
df14.groupby(['month','category_title']).sum()['video_id'].unstack().plot(ax=ax)
ax.set_xlabel('Month')
ax.set_ylabel('Number of videos of particular category')
```
