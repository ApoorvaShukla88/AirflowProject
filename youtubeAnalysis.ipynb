```python
import pandas as pd
import calendar
import json
```

```python
df = pd.read_csv('/Users/amishra/DEV/DataEngineering.Labs.AirflowProject/DataEngg-Airflow/youtube-new/USvideos.csv')
df['month'] = df.apply(lambda row: calendar.month_name[int(row.trending_date.split(".")[2])], axis=1)
df
```

```python
#df1 = pd.read_json('/Users/amishra/DEV/DataEngineering.Labs.AirflowProject/DataEngg-Airflow/youtube-new/US_category_id.json')
#df1.items

data = None

with open('/Users/amishra/DEV/DataEngineering.Labs.AirflowProject/DataEngg-Airflow/youtube-new/US_category_id.json') as f:
data = json.load(f)

category_list = list()
for c in data['items']:
    category_list.append([c['id'],c['snippet']['title']])

df11 = pd.DataFrame(category_list, columns = ['category_id', 'category_title'])
df11
```
